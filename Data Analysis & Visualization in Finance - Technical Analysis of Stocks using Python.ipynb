```python
def trend(x):
  if x > -0.5 and x <= 0.5:
    return 'Slight or No change'
  elif x > 0.5 and x <= 1:
    return 'Slight Positive'
  elif x > -1 and x <= -0.5:
    return 'Slight Negative'
  elif x > 1 and x <= 3:
    return 'Positive'
  elif x > -3 and x <= -1:
    return 'Negative'
  elif x > 3 and x <= 7:
    return 'Among top gainers'
  elif x > -7 and x <= -3:
    return 'Among top losers'
  elif x > 7:
    return 'Bull run'
  elif x <= -7:
    return 'Bear drop'
HDFC_df['Trend']= np.zeros(HDFC_df['Day_Perc_Change'].count())
HDFC_df['Trend']= HDFC_df['Day_Perc_Change'].apply(lambda x:trend(x))
HDFC_df.head()
```

```python
DPC_pie_data = HDFC_df.groupby('Trend')
pie_label = sorted([i for i in HDFC_df.loc[:, 'Trend'].unique()])
plt.pie(DPC_pie_data['Trend'].count(), labels = pie_label, 
       autopct = '%1.1f%%', radius = 2)
plt.show()
```

```python
import pandas_datareader.data as web

start = datetime.datetime(2019, 6, 6)
end = datetime.datetime(2020, 5, 6) 

extracted_df = web.DataReader(['HDFC.NS', 'JINDALSTEL.NS',    'JUBLFOOD.NS','SUNPHARMA.NS', 'TCS.NS', '^NSEI'],
'yahoo', start = start, end = end)['Adj Close']
```

```python
extracted_df.dropna(inplace = True, axis = 0)
extracted_df.head()
```

```python
pct_chg_df = extracted_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = 'any', axis = 0)

sns.set(style = 'ticks', font_scale = 1.25)
sns.pairplot(pct_chg_df)
```

```python
from scipy.stats import stats
```

```python
sns.jointplot('SUNPHARMA.NS', 'JINDALSTEL.NS', pct_chg_df, kind = 'scatter').annotate(stats.pearsonr)
sns.jointplot('JINDALSTEL.NS', 'HDFC.NS', pct_chg_df, kind = 'scatter').annotate(stats.pearsonr)
plt.show()
```

```python
HDFC_volality = pct_chg_df['HDFC.NS'].rolling(7).std()*np.sqrt(7)
HDFC_volality.plot(figsize = (15, 7))
```

```python
volality = pct_chg_df[['HDFC.NS', 'SUNPHARMA.NS', '^NSEI']].rolling(7).std()*np.sqrt(7)
volality.plot(figsize = (15, 7))
```
